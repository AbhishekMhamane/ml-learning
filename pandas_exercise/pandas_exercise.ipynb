```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', None, 'Alice', 'Bob'],
    'Age': [25, None, 30, 28, 25, None],
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'New York', 'Los Angeles']
}
df = pd.DataFrame(data)
df.head()
```

```python
# drop null values
df.dropna()
```

```python
# drop duplicates
df.drop_duplicates(subset=['Name', 'City'])
```

```python
df.shape
```

```python
df.info()
```

```python
# take unique values count
df['Name'].value_counts(dropna=True)
```

```python
df.notnull()
```
